## GPT로 한국어 문장 생성

```python
import numpy as np
import transformers
from transformers import pipeline, set_seed
```

```python
generator = pipeline('text-generation', model='skt/ko-gpt-trinity-1.2B-v0.5')
```

```python
set_seed(42)
```

```python
completion = generator("우리 나라에서 자연어 처리를",
                       max_length=30,          # 문장 생성 최대 길이
                       do_sample=True,
                       pad_token_id=2,
                       num_return_sequences=1, # 하나의 생성 결과만 반환
                       truncation=True)
```

```python
print(completion[0]['generated_text'])   # 첫 번째 메모리 영역에서 'generated_text' 항목에서 결과를 확인할 수 있음
```
