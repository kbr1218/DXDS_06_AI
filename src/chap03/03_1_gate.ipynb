## ✅퍼셉트론
- 인공 뉴런, 단순 퍼셉트론
- 다수의 신호를 입력으로 받아 하나의 신호를 출력
- 신호: 전류나 강물처럼 흐름이 있는 것을 상상
- 퍼셉트론 신호도 흐름을 만들고 정보를 앞으로 전달
- 퍼셉트론 신호는 ‘흐른다(1) / 안 흐른다(0)’의 두 가지 값으로 표현

<br>

- 입력으로 두 개의 신호를 받는 퍼셉트론의 예
    - 
    - $x1$/ $x2$: 입력 신호
    - $y$: 출력 신호
    - $w1$/$w2$: 가중치(weight)
        - 각 신호가 결과에 주는 영향력을 조절하는 요소로 작용
    - 뉴런( | 노드)
    - $theta$: 임계값 (뉴런의 활성화 기준값)

<br>

- 단순 논리 회로
    - **AND** Gate: 두 값이 모두 1일 때 1
    - **NAND** Gate: AND Gate의 반대 값
    - **OR** Gate: 둘 중 하나라도 1일 때 1

<br>

- 다층 퍼셉트론: **XOR** Gate
    - 단순 논리회로는 구현할 수 없기 때문에 기존 게이트의 조합으로 구현 가능
    
    - 0층의 두 뉴런이 입력 신호를 받아 1층의 뉴런으로 신호를 보냄
    - 1층의 뉴런이 2층의 뉴런으로 신호를 보내고, 2층의 뉴런은 이 입력신호를 바탕으로 $y$ 출력

# 03-1. Gate

```python
import tensorflow as tf
import numpy as np
```

<hr> 

### and 연산자

```python
# Gate
x_data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=np.float32)

# AND gate
y_data = np.array([[0], [0], [0], [1]], dtype=np.float32)
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units = 1,
                                input_dim=2,
                                activation='sigmoid'))
model.compile(loss='binary_crossentropy',
             optimizer=tf.optimizers.SGD(learning_rate=0.01),
             metrics=['accuracy'])
model.summary()
```

```python
loss = model.fit(x_data, y_data, epochs=4000)
```

```python
print(model.layers[0].get_weights())
```

```python
# 예측 모델에 x_data 다시 넣기
model.predict(x_data)
```

```python
# 검증 메소드
score = model.evaluate(x_data, y_data)
print('정확도:', score[1])
```
