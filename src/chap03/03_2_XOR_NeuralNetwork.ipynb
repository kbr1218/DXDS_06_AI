# XOR Neural Network

```python
import tensorflow as tf
import numpy as np
```

```python
# Gate
x_data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=np.float32)

# AND gate
y_data = np.array([[0], [1], [1], [0]], dtype=np.float32)
```

```python
model = tf.keras.Sequential()

# 다층 레이어 구성하기 (맨 처음 노드의 입력과 맨 마지막 노드의 출력 노드 수만 정해져있으면 됨)
model.add(tf.keras.layers.Dense(units = 2, input_dim=x_data.shape[1], activation='sigmoid'))  # input_dim : 입력 데이터의 feautre

# 앞의 레이어 units(출력)와 뒷 레이어의 입력(input_dim)이 같아야 함 
# 레이어를 연결하면 연결되는 시그모이드도 늘어남 >> SVM의 gamma값처럼 그래프의 곡률이 높아짐
model.add(tf.keras.layers.Dense(units = 1, activation='sigmoid')) 

model.compile(loss='binary_crossentropy',
             optimizer=tf.optimizers.SGD(learning_rate=0.1),
             metrics=['accuracy'])
model.summary()
```

```python
loss = model.fit(x_data, y_data, epochs=7000)
```

```python
print(model.layers[0].get_weights())
```

```python
print(model.layers[1].get_weights())
```

```python
model.predict(x_data)
```

```python
score = model.evaluate(x_data, y_data)
print('acc:', score[1])
```
