# 02-3. Softmax Classifier

- 100마리에 동물에 대한 16개의 feature

```python
import tensorflow as tf
import numpy as np
```

```python
data = np.loadtxt('data/data-04-zoo.csv', delimiter=',',
                 dtype=np.float32)

x_data = data[:, :-1]
y_data = data[:, [-1]]
```

```python
print(x_data.shape)
print(y_data.shape)

# DV가 가지는 카테고리
nb_classes = 7
# IV의 feature 개수
feature_num = x_data.shape[1]
```

```python
# 7개의 카테고리를 가지는 DV(y_data)와 카테고리수를 넣어 one-hot encoding
y_one_hot = tf.keras.utils.to_categorical(y_data, nb_classes)
y_one_hot[:5]
```

- 다중 분류(softmax)에서 제일 먼저 DV의 값을 확인
- 레이블로 되어있다면 one-hot-encoding으로 0/1 값으로 변환
- 원-핫 인코딩을 하는 이유
    - 정답을 각각의 레이블 확률로 표현하기 위해 (e.g. 0.1, 0.1, 0.7, 0.05, 0.05, ..)

```python
model = tf.keras.Sequential()
```

```python
# 입력 데이터 16개, 출력 데이터 7개
model.add(tf.keras.layers.Dense(units=nb_classes,  # 출력값의 개수
                                input_dim=feature_num
                                 # activation='softmax'   # 아래 코드 대신 작성하는 방법
                               ))
# 다중분류에서 활성화 함수: softmax
# cf> 선형회귀: 'linear', 로지스틱:'sigmoid'
model.add(tf.keras.layers.Activation('softmax'))
```

```python
model.compile(loss='categorical_crossentropy',  # 손실함수- 다중분류: categorical_crossentropy Error (CEE)
                                                # cross-entropy error: 모델이 예측한 확률값과 정답 데이터의 확률값의 분포(각 레이블 값과 예측한 확률값의 차이)에 대한 오차 계산
                                                #                      이 오차가 최소화가 될 때의 w와 b값을 찾는 것
                                                # cf> 선형회귀: MSE (선형함수에만 적용 가능)
                                                #     로지스틱: binary_crossentropy
              optimizer=tf.keras.optimizers.SGD(learning_rate=0.1),
              # 경사하강법의 learning_rate을 0.01로 할 경우 optimizer='sgd'
              metrics=['accuracy']
             )
```

```python
model.summary()
```

- Param (파라미터의 수) >> `119`개 (회귀계수와 bias의 수)
- IV개수 16개 X DV 7개 + bias 7개 = 119개

```python
loss = model.fit(x_data,
                 y_one_hot,
                 epochs=1000)
```

```python
# x_data의 70번째 값을 넣어 테스트 해보기
print(x_data[70])
pred = model.predict(x_data[[70]])
pred
```

```python
# 예측 결과에서 가장 큰 값의 인덱스를 찾는 방법
pred.argmax(axis=1)
```
