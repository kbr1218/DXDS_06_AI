# 05. SVM (Support Vector Machine; 서포트 벡터 머신)

<br><hr>

## ✅서포트 벡터머신
- 사용하기 편하면서도 높은 정확도를 보이는 데이터 분류를 위한 지도학습 머신러닝 알고리즘
- 서포트 벡터를 사용해서 결정 경계를 정의하고, 분류되지 않은 점을 해당 결정 경계와 비교해서 분류

<br>

- SVM의 중요 용어
    - **결정경계 (Decision Boundary)** : 서로 다른 분류 값을 결정하는 경계
    - **서포트 벡터 (Support Vector)** : 결정 경게선과 가장 가까이 맞닿은 데이터 포인트(클래스의 점들)
    - **마진 (margin)** : 서포트 벡터와 결정경계 사이의 거리

<br>

- **SVM의 목표는 마진을 최대로 하는 결정경계를 찾는 것**

<br><hr>

## ✅결정경계 (Decision Bounndary)
- 데이터에 2개 속성(feature)만 있다면 결정 경계는 **선** 형태
- ![svm](img/svm1.png)

<br>

- 속성이 세 개로 늘어난다면 3차원 **평면** 이 됨
- ![svm](img/svm2.png)
    - 이 때의 결정 경계는 '선'이 아닌 '평면'
    - 우리가 이렇게 시각적으로 인지할 수 있는 범위는 3차원까지임
    - 차원, 즉 속성의 개수가 늘어날수록 당연히 복잡
    - 결정경계도 단순한 평면이 아닌 고차원이 될 텐데 이를 **'초평면 hyperplane)'** 이라고 함
    
<br>
    
- **최적의 결정경계**
    - 데이터군으로부터 최대한 멀리 떨어져 있는 게 좋음
    - **Support Vectors** 는 결정경계와 가까이 있는 데이터포인트를 의미
    
<br><hr>

## ✅마진 (Margin)
- **결정경계와 서포트 벡터 사이의 거리** 의미
- 최적의 결정경계는 **마진을 최대화** 함
- n개의 속성을 가진 데이터에는 **최소 n+1개의 서포트 벡터가 존재**
    
<br><hr>

## ✅커널 트릭 (kernel trick)
- 매핑 함수를 가지고 실제로 많은 양의 데이터를 저차원에서 고차원으로 옮기기에는 계산량이 너무 많아서 현실적으로 사용하기 어려움
- 데이터를 고차원으로 보내진 않지만 보낸 것과 동일한 효과를 줘서 매우 빠른 속도로 결정 경계선을 찾는 방법
- 저차원에서 결정경계를 찾지 못할 때 고차원으로 데이터(벡터)를 옮겨서 결정경계를 찾는 방법

<br><hr>

## ✅파라미터 튜닝
- **결정경계** : N-1 차원 (단, 데이터의 벡터 공간은 N차원)
    - 따라서, 2차원 공간의 결정경계는 선, 3차원 공간의 결정경계는 면
- **파라미터 C (Cost; 비용)** : 허용되는 오류 양
    - 마진의 너비를 조정하는 파라미터
    - C 값이 클수록 마진은 낮아지고, 학습 에러율은 감소하는 방향으로 결정 경계선을 만들고, 이를 **하드 마진(hard margin)** 이라고 함
        - 너무 높으면 과대적합의 위험이 있음
    - C 값이 작을수록, 마진을 최대한 높이고, 학습 에러율은 증가시키는 방향으로 결정 경계선을 만들어서 **소프트 마진(soft margin)** 을 만듦
        - 너무 낮으면 과소적합의 위험이 있음

<br>

- SVM에서는 선형으로 분리할 수 없는 점들을 분류하기 위해 **커널(kernel)** 사용
- 커널은 원래 가지고 있는 데이터를 더 높은 차원의 데이터로 변환
    - 2차원의 점으로 나타낼 수 있는 데이터를 **다항식(polynomial)** 커널은 3차원으로,
    - **RBF(Radial Basis Function)** 은 커널은 점을 무한한 차원으로 변환

<br>

- RBF 커널에는 파라미터 **감마(gamma)** 가 있음
    - 커널의 데이터포인트 표준 편차를 결정하는 조절변수
    - 감마가 너무 크면 학습 데이터테 너무 의존해서 overfitting이 발생할 수 있음

<br>

- 대부분의 머신러닝 지도학습 알고리즘은 학습 데이터 모두를 사용하여 모델을 학습
    - SVM에서는 결정경계를 정의하는 게 결국 서포트 벡터이기 때문에, 데이터 포인트 중에서 서포트 벡터만 잘 골라내면 나머지 쓸 데 없는 수많은 데이터 포인터들을 무시할 수 있음
    - 그래서 **매우 빠름**

<br><hr>

## ✅SVM 장/단점
#### 장점
1. 커널 트릭을 사용함으로써 특성이 다양한 데이터를 분류하는 데 강함
    - N개의 특성을 가진 데이터는 N차원 공간의 데이터 포인트로 표현되고, N차원 공간 또는 그 이상의 공간에서 초평면을 찾아 데이터 분류가 가능하기 때문
2. 예측 속도가 빠름
3. 파라미터(C, gamma)를 조정해서 과대적합 및 과소적합에 대처할 수 있음
4. 적은 학습 데이터로도 딥러닝만큼 정확도가 높은 분류를 기대할 수 있음

<br>

#### 단점
1. 데이터 전처리 과정이 상당히 중요
    - 특성이 비슷한 수치로 구성된 데이터 같은 경우에는 SVM을 쉽게 활용할 수 있지만, 특성이 다양하거나 혹은 확연히 다른 경우에는 데이터 전처리 과정을 통해 데이터 특성 그대로 벡터 공간에 표현해야 함
2. 특성이 많을 경우 결정경계 및 데이터의 시각화가 어려움
    - SVM의 분류 결과를 이해하기 힘듦
3. 커널 트릭 오사용 시 과대적합되기 쉬움

<br><hr>

## 00. 기본 설정

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

## 01. 데이터 가져오기

```python
train = pd.read_csv('data/train.csv', encoding='cp949')
test = pd.read_csv('data/test.csv', encoding='cp949')
```

```python
# shape 확인
print("train shape", train.shape)
print("test shape", test.shape)
```

<hr>

## 02. 최적의 파라미터 찾기
##### 1. C(cost; 비용): 결정경계선의 마진을 결정하는 파라미터
##### 2. gamma: 커널의 데이터포인트 표준편차를 결정하는 조절 변수

```python
# C값이 커지면 마진의 값이 작아짐
# gamma가 커지면 결정경계선이 직선에서 원형에 가까운 곡률의 형태로 결정경계선이 만들어짐

from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC # support vector classifier
```

```python
# sklearn에서 제공하는 GridSearchCV를 사용하면, 손쉽게 cost와 gamma 값을 찾을 수 있음

def svm_param_selection(X, y, nfolds):
    svm_parameters = [{
        'kernel':['rbf'],
        'C':[0.01, 0.1, 1, 10, 100, 1000],
        'gamma':[0.00001, 0.0001, 0.001, 0.01, 0.1, 1]        
    }]
    
    clf = GridSearchCV(SVC(), svm_parameters, cv=nfolds)
    clf.fit(X, y)
    print(clf.best_params_)
    
    return clf
```

```python
# 두 개의 feature만 가지고 모델학습
X_train = train[['3P', 'BLK']]
y_train = train[['Pos']]
```

```python
clf = svm_param_selection(X_train, y_train.values.ravel(), 10)
```

<hr>

## 03. 시각화
##### 최적의 파라미터일 때의 결정경계선과 다른 파라미터들일 때의 결정경계선을 비교

```python
# 시각화를 하기 위해 최적의 C와 최적의 C와 비교하기 위한 다른 C 후보값으로 저장
C_candidate = []
C_candidate.append(clf.best_params_['C'] * 0.01)
C_candidate.append(clf.best_params_['C'])
C_candidate.append(clf.best_params_['C'] * 100)
```

```python
# 시각화를 하기 위해 최적의 gamma와 최적의 gamma와 비교하기 위한 다른 gamma 후보값으로 저장
gamma_candidate = []
gamma_candidate.append(clf.best_params_['gamma'] * 0.01)
gamma_candidate.append(clf.best_params_['gamma'])
gamma_candidate.append(clf.best_params_['gamma'] * 100)
```

```python
X = train[['3P', 'BLK']]
Y = train['Pos'].tolist()
```

```python
# 포지션에 해당하는 문자열 SG와 C를 벡터화
position = []

for i in Y:
    if i == 'C':
        position.append(0)
    else:
        position.append(1)
        
# 각각의 파라미터에 해당하는 SVM 모델을 만들어 classifiers에 저장
classifiers = []             # 분류기를 저장하기 위한 리스트 선언

for C in C_candidate:
    for gamma in gamma_candidate:
        # 총 3*3 = 9개의 모델 생성
        clf = SVC(C=C, gamma=gamma)
        clf.fit(X, Y)                # 학습 데이터와 정답 데이터를 가지고 학습
        classifiers.append((C, gamma, clf))
```

```python
# 18,18 사이즈의 차트를 구성합니다.
plt.figure(figsize=(18, 18))
xx, yy = np.meshgrid(np.linspace(0, 4, 100), np.linspace(0, 4, 100))

# 각각의 모델들에 대한 결정 경계 함수를 적용하여 함께 시각화
for (k, (C, gamma, clf)) in enumerate(classifiers):
    # enumerate: 데이터를 인덱스와 함께 반환
    Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # 최적의 모델을 포함한 다른 파라미터로 학습된 모델들을 함께 시각화해봅니다.
    plt.subplot(len(C_candidate), len(gamma_candidate), k + 1)
    plt.title("gamma=10^%d, C=10^%d" % (np.log10(gamma), np.log10(C)),
              size='medium')

    # 서포트 벡터와 결정경계선을 시각화합니다.
    plt.pcolormesh(xx, yy, -Z, cmap=plt.cm.RdBu) # 2D 배열 플로팅(cmap="plasma")
    plt.scatter(X['3P'], X['BLK'], c=position, cmap=plt.cm.RdBu_r, edgecolors='k')
```

```python
# 결정경계는 원래 선형으로 그려지지만, gamma값이 커지면 직선>>곡선을 그림
```

```python
# 테스트에 사용될 feature 지정
X_test = test[['3P', 'BLK']]
y_test = test[['Pos']]

pred = clf.predict(X_test)
pred
```
