## ✅선형 회귀분석 (Linear Regression Analysis)
- 두 개 또는 그 이상의 변수 간 인과관계 패턴을 원래 모습과 가장 가깝게 추정하는 분석 방법
- 함수 $y=2x + 80$의 그래프, 이를 선형 회귀분석의 모형(Model)이라고 부름
- 선형 회귀분석은 x변수가 원인, y변수가 결과로 인과관계여야 한다는 조건이 있음
 ![LR](img/linearRegression1.png)

<br>
<hr>

## ✅선형 회귀분석의 모형
- 회귀분석에서 원인인 x 변수는 독립변수(IV; Independent Variable), 결과인 y 변수는 종속변수(DV; Dependent Variable)라고 부름
    - **단순 선형 회귀분석(Simple Linear Regression Analysis)**: 어떠한 결과의 원인이 되는 독립변수가 한 개일 때
    - **다중 선형 회귀분석(Multiple Linear Regression Analysis)**: 독립변수가 두 개 이상일 때

<br>

- 공부 시간과 시험 점수의 인과관계를 분석해보면 종속변수는 시험 점수이며 독립변수는 공부 시간만 고려
     ![LR2](img/linearRegression2.png)
    
<br>

- 직선을 일차함수 $y=mx+b$로 표현한다면 계수 m은 이 직선의 기울기이므로 양수일 것
- 일반적인 선형회귀 모형은 다음 식과 같이 독립변수 $x_i$ 앞에 계수 $\beta_i$가 붙음
- 마지막 $\epsilon$는 오차항
$$ y = \beta_{0} + \beta_{1}x_{1} + \beta_{2}x_{2} + \beta_{3}x_{3} + ... + \epsilon $$

- 데이터를 표현하는 직선은 여러 개 존재할 수 있음
     ![LR2](img/linearRegression3.png)

<br>

- 최대한 많은 점과 거리가 가까운 직선이 좋은 직선
- 점에서 직선까지 y축과 평행한 선분을 그렸을 때 모든 선분 길이의 합을 최소로 하는 직선을 찾는 것
     ![LR2](img/linearRegression4.png)
- 점이 가리키는 값과 직선이 예측하는 값의 차를 **잔차(Residual)** 라고 부름

<br>
<hr>

## ✅ 더미변수 (Dummy Variable)
- 독립변수를 0과 1로 변환하여 '예'와 '아니오'로 나타낼 수 있는 변수
- 더미 변수를 여러 개 두면 '예'와 '아니오'만으로 결과를 세 가지 이상으로 나눌 수 있음
- 구분하고자 하는 데이터의 종류가 N개일 때 더미변수 N-1개를 선언

<br>
<hr>

## ✅ 결정계수
- 선형 회귀분석에서 모형이 데이터의 패턴을 얼마나 효과적으로 보여주는지 수치화한 값
- 결정계수 **R2(R square)** 를 아래 수식과 같이 정의
     ![LR2](img/linearRegression5.png)
     
<br>

- 편차(Deviation)는 **평균과 실제값의 차이**
- 결정계수 $R^2$는 0 이상 1 이하의 값으로 계산됨
    - $R^2$ 값이 1에 가까울 때 잔차가 작고 예측의 정밀도가 높음
    - $R^2$ 값이 0에 가까울 때 잔차가 커 예측의 정밀도가 낮음
     ![LR2](img/linearRegression6.png)

<br>
<hr>

## ✅ 수정된 결정계수
- 결정계수 $R^2$는 독립변수의 개수가 많을수록 커지는 경향을 보임
- 이 문제를 해결하기 위해 다중 선형회귀 분석에서 **수정된 결정계수 adj.R2 (adjusted R Square)** 로 설명력을 나타냄
     ![LR2](img/linearRegression7.png)
     
<br>
<hr>

## ✅ 선형 회귀분석의 해석
- 통계적 가설검정
    - 모집단에 대한 추축을 하고 표본의 정보를 기준으로 가설이 타당한지 판정하는 방법
    - 통계적 가설에는 두 종류가 있음
        - **귀무가설(Null Hypothesis)**: 통계학에서 처음부터 거짓일 것으로 기대하는 가설
        - **대립가설(Alternative Hypothesis)**: 입증하고자하는 가설
    - 실험 결과를 보고 귀무가설을 채택하거나 대립가설을 채택하는 기준을 세워두어야 하고, 그 기준을 **유의수준(Significance Level)**이라고 함
    - 귀무가설이 참일 때 실제 결과가 실험 결과와 같을 확률은 **p-값(p-value; 유의확률)**이라고 함
    
<br>

- 선형 회귀분석 과정
    - 선형 회귀분석을 수행하고 해석하는 과정
        1. 결과인 종속변수를 y로 두고, 원인이 되는 독립변수를 x로 둠
        2. 설명력 $R^2$ 또는 $adj.R^2$ 값을 확인 (결정계수가 0.6 또는 0.4 이상이면 해당 회귀모형이 설명력을 갖추었다고 인정
        3. 각 독립변수의 p-value가 유의수준보다 작은지 확인. p-value가 유의수준 이상인 변수를 제외하고 남은 독립변수가 결과에 영향을 주는 원인

## 01. 데이터 불러오기

```python
# 연봉 (만 원)
x = [3000, 4200, 4000, 5000, 6000, 3800, 3500, 6200, 3900, 4500]
# 직장 만족도 (점)
y = [60, 75, 70, 85, 90, 70, 65, 95, 70, 80]
```

```python
# 데이터프레임으로 생성
data = {'x':x, 'y':y}
df = pd.DataFrame(data)
df
```

```python
sns.scatterplot(data=df, x='x', y='y')
```

```python
plt.close('all')
```

<hr>

## 02. 선형회귀분석

```python
from statsmodels.formula.api import ols
from sklearn.linear_model import LinearRegression
```

```python
fit = ols('y ~ x', data=df).fit()
fit.summary()
```

```python
# 데이터의 설명력 R2 >> 0.6을 초과하므로 모형의 정밀도가 높다고 할 수 있음
fit.rsquared
```

```python
# x의 유의수준 0.000 >> x는 유의한 독립변수 
fit.pvalues
```

## 03. 다중선형회귀

#### 직장 만족도의 요인 분석

```python
# 직장 만족도 (점)
y = [60, 75, 70, 85, 90, 70, 65, 95, 70, 80]
# 연봉 (만 원)
x1 = [3000, 4200, 4000, 5000, 6000, 3800, 3500, 6200, 3900, 4500]
# 일평균 휴식시간 (분)
x2 = [120, 60, 100, 100, 50, 120, 90, 40, 120, 120]
# 일평균 근무시간 (시간)
x3 = [8, 6, 10, 8, 10, 10, 9, 7, 8, 9]
```

```python
# 데이터프레임으로 생성
data = {'satisfaction':y, 'salary':x1, 'breakh':x2, 'workhour':x3}
df = pd.DataFrame(data)
df
```

```python
fit = ols('satisfaction ~ salary + breakh + workhour', data=df).fit()
fit.summary()
```

```python
# 각각 변수별 pvalue 확인
fit.pvalues
```
